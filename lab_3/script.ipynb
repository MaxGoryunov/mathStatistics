```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def Normal(n):
    return np.random.normal(0, 1, n)

def Cauchy(n):
    return np.random.standard_cauchy(n)

def Student(n):
    return np.random.standard_t(3, n)

def Poisson(n):
    return np.random.poisson(10, n)

def Uni(n):
    return np.random.uniform(-math.sqrt(3), math.sqrt(3), n)
```

```python
labels = [
  ("Normal", Normal),
  ("Cauchy", Cauchy),
  ("Student", Student),
  ("Poisson", Poisson),
  ("Uni", Uni),
]
```

```python
sizes = [20, 100]
```

```python
for label, func in labels:
  figure, axes = plt.subplots(2, 1, figsize=(20, 10))

  for index, size in enumerate(sizes):
    values = func(size)

    sns.boxplot(values, ax=axes[index], orient="h")
    axes[index].set_xlabel("x")
    axes[index].set_title(f"n = {size}")

  figure.savefig(f"plots/{label}_boxplot.png")
```
